```python
import pickle
import os
import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread, imshow, imsave
```

```python
def display(image,title="Image"):
    plt.figure(figsize=[10,10])
    channels=len(image.shape)
    if channels<3:
        plt.imshow(image,cmap='Greys_r');
    else:
        plt.imshow(image);      
    plt.title(title);plt.axis("off");  
```

```python
loaded_model = pickle.load(open('model.pkl', 'rb'))
```

```python
directory_search="C:/Users/saqib/Desktop/Defect Driver"
for filename in os.listdir(directory_search):
    f = os.path.join(directory_search, filename)
    if os.path.isfile(f):
        loaded_model.predict(f, confidence=40, overlap=30).save("prediction.jpg")
        image = imread("prediction.jpg")
        display(image,title="Image")
```
